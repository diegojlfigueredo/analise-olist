```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
covid = pd.read_csv(url, parse_dates=['date'])
covid.head()
```

```python
# Filtrando apenas o Brasil
br = covid[covid['location'] == 'Brazil']

# Conferindo as primeiras linhas e as colunas disponÃ­veis
print(br.shape)
br[['date','total_cases','new_cases','total_deaths','new_deaths']].head()
```

```python
print("PerÃ­odo disponÃ­vel no dataset:", br['date'].min(), "atÃ©", br['date'].max())

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(br['date'], br['new_cases'].rolling(7).mean(), label='MÃ©dia mÃ³vel 7 dias de novos casos')
plt.title('MÃ©dia MÃ³vel 7 dias â€“ Novos Casos de COVID-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('Casos (mÃ©dia mÃ³vel)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("Maior nÃºmero diÃ¡rio de casos:", br['new_cases'].max())
print("Data do pico de casos:",
      br.loc[br['new_cases'].idxmax(), 'date'])
print("Maior nÃºmero diÃ¡rio de mortes:", br['new_deaths'].max())
print("Data do pico de mortes:",
      br.loc[br['new_deaths'].idxmax(), 'date'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(br['date'], br['new_cases'].rolling(7).mean(), label='Casos (mÃ©dia mÃ³vel 7d)')
plt.plot(br['date'], br['new_deaths'].rolling(7).mean(), label='Mortes (mÃ©dia mÃ³vel 7d)')
plt.title('MÃ©dia MÃ³vel de Novos Casos e Mortes de COVID-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('NÃºmero (mÃ©dia mÃ³vel)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(br['date'], br['people_vaccinated'].rolling(7).mean(),
         label='Pessoas vacinadas (mÃ©dia mÃ³vel 7d)')
plt.title('Campanha de VacinaÃ§Ã£o contra COVID-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('Pessoas vacinadas')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculando mÃ©dias mÃ³veis de 7 dias
br['casos_7d']   = br['new_cases'].rolling(7).mean()
br['mortes_7d']  = br['new_deaths'].rolling(7).mean()
br['vacinados_7d'] = br['people_vaccinated'].rolling(7).mean()

plt.figure(figsize=(12,6))

# Eixo casos e mortes
plt.plot(br['date'], br['casos_7d'],  label='Casos (7d)',   linewidth=2)
plt.plot(br['date'], br['mortes_7d'], label='Mortes (7d)',  linewidth=2)

# Eixo vacinaÃ§Ã£o no mesmo grÃ¡fico, mas escala secundÃ¡ria
ax1 = plt.gca()
ax2 = ax1.twinx()
ax2.plot(br['date'], br['vacinados_7d'], label='Vacinados (7d)', linestyle='--')

# Legendas e formataÃ§Ã£o
lines, labels = ax1.get_lines()+ax2.get_lines(), [l.get_label() for l in ax1.get_lines()]+[l.get_label() for l in ax2.get_lines()]
ax1.legend(lines, labels, loc='upper left')
ax1.set_xlabel('Data')
ax1.set_ylabel('Casos / Mortes (mÃ©dia mÃ³vel)')
ax2.set_ylabel('Pessoas Vacinadas (mÃ©dia mÃ³vel)')
plt.title('MÃ©dias MÃ³veis de Casos, Mortes e VacinaÃ§Ã£o â€“ COVID-19 no Brasil')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
