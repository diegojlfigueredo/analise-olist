```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
covid = pd.read_csv(url, parse_dates=['date'])
covid.head()
```

```python
# Filtrando apenas o Brasil
br = covid[covid['location'] == 'Brazil']

# Conferindo as primeiras linhas e as colunas disponíveis
print(br.shape)
br[['date','total_cases','new_cases','total_deaths','new_deaths']].head()
```

```python
print("Período disponível no dataset:", br['date'].min(), "até", br['date'].max())

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(br['date'], br['new_cases'].rolling(7).mean(), label='Média móvel 7 dias de novos casos')
plt.title('Média Móvel 7 dias – Novos Casos de COVID-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('Casos (média móvel)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("Maior número diário de casos:", br['new_cases'].max())
print("Data do pico de casos:",
      br.loc[br['new_cases'].idxmax(), 'date'])
print("Maior número diário de mortes:", br['new_deaths'].max())
print("Data do pico de mortes:",
      br.loc[br['new_deaths'].idxmax(), 'date'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(br['date'], br['new_cases'].rolling(7).mean(), label='Casos (média móvel 7d)')
plt.plot(br['date'], br['new_deaths'].rolling(7).mean(), label='Mortes (média móvel 7d)')
plt.title('Média Móvel de Novos Casos e Mortes de COVID-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('Número (média móvel)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(br['date'], br['people_vaccinated'].rolling(7).mean(),
         label='Pessoas vacinadas (média móvel 7d)')
plt.title('Campanha de Vacinação contra COVID-19 no Brasil')
plt.xlabel('Data')
plt.ylabel('Pessoas vacinadas')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculando médias móveis de 7 dias
br['casos_7d']   = br['new_cases'].rolling(7).mean()
br['mortes_7d']  = br['new_deaths'].rolling(7).mean()
br['vacinados_7d'] = br['people_vaccinated'].rolling(7).mean()

plt.figure(figsize=(12,6))

# Eixo casos e mortes
plt.plot(br['date'], br['casos_7d'],  label='Casos (7d)',   linewidth=2)
plt.plot(br['date'], br['mortes_7d'], label='Mortes (7d)',  linewidth=2)

# Eixo vacinação no mesmo gráfico, mas escala secundária
ax1 = plt.gca()
ax2 = ax1.twinx()
ax2.plot(br['date'], br['vacinados_7d'], label='Vacinados (7d)', linestyle='--')

# Legendas e formatação
lines, labels = ax1.get_lines()+ax2.get_lines(), [l.get_label() for l in ax1.get_lines()]+[l.get_label() for l in ax2.get_lines()]
ax1.legend(lines, labels, loc='upper left')
ax1.set_xlabel('Data')
ax1.set_ylabel('Casos / Mortes (média móvel)')
ax2.set_ylabel('Pessoas Vacinadas (média móvel)')
plt.title('Médias Móveis de Casos, Mortes e Vacinação – COVID-19 no Brasil')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filtrar e criar cópia
br = covid[covid['location'] == 'Brazil'].copy()

# Agora crie as colunas de média móvel sem warning
br['casos_7d']     = br['new_cases'].rolling(7).mean()
br['mortes_7d']    = br['new_deaths'].rolling(7).mean()
br['vacinados_7d'] = br['people_vaccinated'].rolling(7).mean()
```
